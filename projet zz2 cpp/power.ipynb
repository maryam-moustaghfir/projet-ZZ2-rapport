```python
import matplotlib.pyplot as plt

# Chemin vers le fichier texte
fichier = "powerft10_9.txt"

# Initialisation des listes pour stocker les indices et les valeurs
indices = []
valeurs = []

# Lecture du fichier texte
with open(fichier, 'r') as f:
    for ligne in f:
        # Séparation de la ligne en deux parties
        partie1, partie2 = ligne.strip().split(", ")
        # Récupération de l'indice et de la valeur
        indice = int(partie1.split(": ")[1])
        valeur = int(partie2.split(": ")[1])
        # Ajout à la liste des indices et des valeurs
        indices.append(indice)
        valeurs.append(valeur)

# Préparation des listes pour la nouvelle courbe
new_indices = []
new_valeurs = []

# Création de la courbe carrée
for i in range(len(valeurs) - 1):
    new_indices.append(indices[i])
    new_valeurs.append(valeurs[i])

    # Si les valeurs adjacentes sont différentes
    if valeurs[i] != valeurs[i + 1]:
        new_indices.append(indices[i + 1])
        new_valeurs.append(valeurs[i])

# Ajout du dernier point
new_indices.append(indices[-1])
new_valeurs.append(valeurs[-1])

# Tracé de la courbe carrée
plt.plot(new_indices, new_valeurs, color='blue', linestyle='-')
plt.axhline(y=500, color='red', linestyle='--', label='Puissance 500')

plt.xlabel('Temps (unités)')
plt.ylabel('Puissance')
plt.title('Courbe de Puissance en fonction du Temps')
plt.grid(True)
plt.show()
```
